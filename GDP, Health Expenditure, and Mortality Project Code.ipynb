```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a list of UN countries

UN_countries = pd.read_csv('/content/un_member_iso_region.csv')

UN_countries = UN_countries.rename(columns={'iso3': 'Country Code'})

UN_countries = UN_countries.rename(columns={'member': 'Country Name'})

UN_countries = UN_countries.drop(columns=['iso2', 'region', 'unga_region_group'])

UN_countries
```

```python
GDP_per_capita = pd.read_csv('/content/API_NY.GDP.PCAP.CD_DS2_en_csv_v2_85121.csv', skiprows=4)
# GDP per Capita for 2022
GDP_per_capita_2022 = GDP_per_capita[
    ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code', '2022']
]

GDP_per_capita_2022
```

```python
Total_GDP = pd.read_csv('/content/API_NY.GDP.MKTP.CD_DS2_en_csv_v2_85078.csv', skiprows=4)
# Total GDP for 2022
Total_GDP_2022 = Total_GDP[
    ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code', '2022']
]

Total_GDP_2022
```

```python
Health_Expenditure_Percentage_of_GDP = pd.read_csv('/content/API_SH.XPD.CHEX.GD.ZS_DS2_en_csv_v2_85318.csv', skiprows=4)
# Health Expenditure Percentage of GDP for 2022
Health_Expenditure_Percentage_of_GDP_2022 = Health_Expenditure_Percentage_of_GDP[
    ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code', '2022']
]

Health_Expenditure_Percentage_of_GDP_2022
```

```python
Health_Expenditure_Per_Capita = pd.read_csv('/content/API_SH.XPD.CHEX.PC.CD_DS2_en_csv_v2_87561.csv', skiprows=4)
# Health Expenditure per Capita for 2022
Health_Expenditure_Per_Capita_2022 = Health_Expenditure_Per_Capita[
    ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code', '2022']
]

Health_Expenditure_Per_Capita_2022
```

```python
Mortality_Rate_Infants = pd.read_csv('/content/API_SP.DYN.IMRT.IN_DS2_en_csv_v2_85070.csv', skiprows=4)
# Infant Mortality Rate for 2022
Mortality_Rate_Infants_2022 = Mortality_Rate_Infants[
    ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code', '2022']
]

Mortality_Rate_Infants_2022
```

```python
Mortality_Rate_Adults = pd.read_csv('/content/Mortality rate, adult (per 1,000 adults).csv')
Mortality_Rate_Adults_2022 = Mortality_Rate_Adults[Mortality_Rate_Adults["Year"] == 2022]
mortality_pivot = Mortality_Rate_Adults_2022.pivot_table(
    index=['Country Code', 'Country Name'],
    columns='Disaggregation',
    values='Value'
).reset_index()

mortality_pivot.columns.name = None
mortality_pivot = mortality_pivot.rename(columns={
    'female': 'Adult_Mortality_Female',
    'male': 'Adult_Mortality_Male',
})

mortality_pivot
```

```python
# Merging all dataframes into a singular dataframe based on country code
merged_df = pd.merge(
    UN_countries[['Country Code', 'Country Name']],
    Total_GDP_2022[['Country Code', '2022']],
    on='Country Code',
    how='inner'
)
merged_df = merged_df.rename(columns={'2022': 'GDP_total'})

merged_df = pd.merge(
    merged_df,
    GDP_per_capita_2022[['Country Code', '2022']],
    on='Country Code',
    how='inner'
)
merged_df = merged_df.rename(columns={'2022': 'GDP_per_capita'})

merged_df = pd.merge(
    merged_df,
    Health_Expenditure_Percentage_of_GDP_2022[['Country Code', '2022']],
    on='Country Code',
    how='inner'
)
merged_df = merged_df.rename(columns={'2022': 'Health_Expenditure_Pct_GDP'})


merged_df = pd.merge(
    merged_df,
    Health_Expenditure_Per_Capita_2022[['Country Code', '2022']],
    on='Country Code',
    how='inner'
)
merged_df = merged_df.rename(columns={'2022': 'Health_Expenditure_Per_Capita'})


merged_df = pd.merge(
    merged_df,
    Mortality_Rate_Infants_2022[['Country Code', '2022']],
    on='Country Code',
    how='inner'
)
merged_df = merged_df.rename(columns={'2022': 'Infant_Mortality_Rate'})


merged_df = pd.merge(
    merged_df,
    mortality_pivot[['Country Code', 'Adult_Mortality_Female', 'Adult_Mortality_Male']],
    on='Country Code',
    how='inner'
)

pd.set_option('display.max_rows', None)
merged_df
```

```python
# Separate features and targets
features = [
    'GDP_per_capita',
    'GDP_total',
    'Health_Expenditure_Pct_GDP',
    'Health_Expenditure_Per_Capita'
]

targets = [
    'Infant_Mortality_Rate',
    'Adult_Mortality_Female',
    'Adult_Mortality_Male'
]

data = merged_df.rename(columns={
    '2022_per_capita': 'GDP_per_capita',
    '2022_total': 'GDP_total'
})

analysis_df = data[features + targets].dropna()
```

```python
# Correlation Matrix
corr = analysis_df[features + targets].corr()
subset_corr = corr.loc[features, targets]

plt.figure(figsize=(3, 3))
ax = sns.heatmap(
    subset_corr,
    annot=True,
    cmap='coolwarm',
    center=0,
    annot_kws={'fontsize': 10, 'fontweight': 'bold'}
)

ax.set_xticklabels(
    [label.get_text().replace('_', ' ') for label in ax.get_xticklabels()],
    rotation=45,
    ha='right',
    fontsize=10,
    fontweight='bold'
)
ax.set_yticklabels(
    [label.get_text().replace('_', ' ') for label in ax.get_yticklabels()],
    rotation=0,
    fontsize=10,
    fontweight='bold'
)

plt.title(
    "Correlation: GDP / Health Spending vs Mortality Metrics",
    fontsize=14,
    fontweight='bold'
)

plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
import math

# Random Forest Regression Model for IMR (GridSearch)

X_train = analysis_df[features]
y_train = analysis_df[targets[0]]


pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', RandomForestRegressor())
])

param_grid = {
    'model__n_estimators': [50, 100, 150],
    'model__max_depth': [None, 10, 20],
}

grid_search = GridSearchCV(
    pipeline,
    param_grid,
    cv=5,
    scoring='neg_mean_squared_error'
)

grid_search.fit(X_train, y_train)

rsme = math.sqrt(-grid_search.best_score_)

print("RSME:", rsme)
print("Best Parameters:", grid_search.best_params_)

importantfeatures = grid_search.best_estimator_.named_steps['model'].feature_importances_
feature_importance_df = pd.DataFrame({'Feature': features, 'Importance': importantfeatures})
print(feature_importance_df)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import math

# Linear Regression Model (GridSearch) for IMF
# Drop GDP_total and Health_Expenditure_PCT_GDP since it shows lower feature importance for Random Forest
features_reduced = [
    'GDP_per_capita',
    'Health_Expenditure_Per_Capita'
]

X_train = analysis_df[features_reduced]
y_train = analysis_df[targets[0]]


pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LinearRegression())
])

param_grid = {
    'model__fit_intercept': [True, False],
    'model__positive': [True, False]
}

grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

print(grid_search.best_params_)

best_model = grid_search.best_estimator_

linreg = best_model.named_steps['model']

coefficients = linreg.coef_

print("Coefficients:")
for feature, coef in zip(features_reduced, coefficients):
    print(f"{feature}: {coef:.4f}")

print("Intercept:", linreg.intercept_)

rsme = math.sqrt(-grid_search.best_score_)

print("RSME:", rsme)
```

```python
import plotly.express as px

# Create health expenditure per capita buckets
merged_df['Health_Spending_Bucket'] = merged_df['Health_Expenditure_Per_Capita'].apply(
    lambda x: 'Low Spending' if x < 500 else 'Medium Spending' if x < 2000 else 'High Spending'
)


df = merged_df[['Health_Spending_Bucket', 'Infant_Mortality_Rate', 'Adult_Mortality_Female', 'Adult_Mortality_Male']].dropna()
df = df.melt(
    id_vars='Health_Spending_Bucket',
    var_name='Mortality Type',
    value_name='Rate'
)

df['Mortality Type'] = df['Mortality Type'].replace({
    'Infant_Mortality_Rate'  : 'Infant',
    'Adult_Mortality_Female' : 'Adult Female',
    'Adult_Mortality_Male'   : 'Adult Male'
})

df = df.groupby(['Mortality Type', 'Health_Spending_Bucket'])['Rate'].mean().reset_index()

fig = px.bar(
    df,
    x='Mortality Type',
    y='Rate',
    color='Health_Spending_Bucket',
    barmode='group',
    width=800,
    height=600,
    color_discrete_sequence=['#2ECC40', '#39CCCC', '#0074D9'],
    title='Mortality Rates by Healthcare Expenditure per Capita Group (2022)',
    labels={'Rate': 'Mortality Rate per 1,000', 'Health_Spending_Bucket': 'Spending Group'},
    category_orders={'Health_Spending_Bucket': ['High Spending', 'Medium Spending', 'Low Spending']}
)

fig.update_layout(
    legend=dict(title='Spending Group', font=dict(size=30)),
    xaxis=dict(title_font=dict(size=25), tickfont=dict(size=19)),
    yaxis=dict(title_font=dict(size=25), tickfont=dict(size=19)),
    title=dict(
        text='Mortality Rates by Healthcare Spending Group',
        x=0.5,
        xanchor='center',
        font=dict(size=30, color='black', weight='bold')
    )
)

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

d = merged_df.dropna(subset=[
    'Country Code', 'Country Name',
    'Health_Expenditure_Per_Capita',
    'Infant_Mortality_Rate'
])

sr = d['Infant_Mortality_Rate'].max() / 900
sizes = np.sqrt(d['Infant_Mortality_Rate'] / sr)

cap = d['Health_Expenditure_Per_Capita'].quantile(.9)

fig = go.Figure()

fig.add_choropleth(
    locations=d['Country Code'],
    z=d['Health_Expenditure_Per_Capita'],
    zmin=0, zmax=cap,
    colorscale='YlGnBu',
    marker_line_color='white',
    colorbar=dict(title='Health Expenditure Per Capita', len=0.55, thickness=12, x=1.04, y=0.40)
)

fig.add_scattergeo(
    locations=d['Country Code'],
    locationmode='ISO-3',
    hovertext=d['Country Name'],
    marker=dict(size=sizes, sizemode='diameter',
                color='rgba(0,0,0,0.55)', line=dict(color='white', width=0.8)),
    showlegend=False
)

for m in [10, 30, 60]:
    fig.add_scattergeo(
        lon=[None], lat=[None],
        marker=dict(size=np.sqrt(m / sr), color='lightgrey', line_color='grey'),
        name=f'IMR ≈ {m}/1,000'
    )

fig.update_layout(
    title=dict(
        text='Health Expenditure pc (color) & Infant mortality rate (bubble size)',
        x=0.5, y=0.8, font=dict(size=25, weight='bold', color = "black")
    ),
    legend=dict(x=1.13, y=0.82, font_size=14, bordercolor='black', borderwidth=0.7, title='Bubble size'),
    geo=dict(showcoastlines=False, showcountries=True, projection_type='natural earth'),
    margin=dict(l=0, r=140, t=60, b=0), width=950, height=800
)

fig.show()
```
